```python
folder_path_2017 = '/Users/thisgirlcan/Desktop/AI_CW2/cw2_data/LIVST_NRCH_OD_a51_2017'
folder_path_2018 = '/Users/thisgirlcan/Desktop/AI_CW2/cw2_data/LIVST_NRCH_OD_a51_2018'
folder_path_2019 = '/Users/thisgirlcan/Desktop/AI_CW2/cw2_data/LIVST_NRCH_OD_a51_2019'
folder_path_2020 = '/Users/thisgirlcan/Desktop/AI_CW2/cw2_data/LIVST_NRCH_OD_a51_2020'
folder_path_2021 = '/Users/thisgirlcan/Desktop/AI_CW2/cw2_data/LIVST_NRCH_OD_a51_2021'
folder_path_2022 = '/Users/thisgirlcan/Desktop/AI_CW2/cw2_data/LIVST_NRCH_OD_a51_2022'

# csv_files = [os.path.join(folder_path_2019, file) for file in os.listdir(folder_path_2019) if file.endswith('.csv')]
# csv_files += [os.path.join(folder_path_2020, file) for file in os.listdir(folder_path_2020) if file.endswith('.csv')]

# csv_files = [os.path.join(folder_path_2017, file) for file in os.listdir(folder_path_2017) if file.endswith('.csv')]
# csv_files = [os.path.join(folder_path_2018, file) for file in os.listdir(folder_path_2018) if file.endswith('.csv')]
csv_files = [os.path.join(folder_path_2019, file) for file in os.listdir(folder_path_2019) if file.endswith('.csv')]
csv_files += [os.path.join(folder_path_2020, file) for file in os.listdir(folder_path_2020) if file.endswith('.csv')]
csv_files += [os.path.join(folder_path_2021, file) for file in os.listdir(folder_path_2021) if file.endswith('.csv')]
csv_files += [os.path.join(folder_path_2022, file) for file in os.listdir(folder_path_2022) if file.endswith('.csv')]
df = pd.concat((pd.read_csv(f) for f in csv_files))

df
```

```python
valid_stations = [
    "LIVST",   
    "DISS",  
    "STWMRKT", 
    "IPSWICH", 
    "MANNGTR", 
    "CLCHSTR", 
    "WITHAME", 
    "CHLMSFD", 
    "INT",     
    "SHENFLD", 
    "STFD", 
    "NRCH",
      ]
df = df[df['tpl'].isin(valid_stations)]
df.head(30)
```

```python
custom_order = [
    "LIVST", "STFD", "SHENFLD", "INT", "CHLMSFD", 
    "WITHAME", "CLCHSTR", "MANNGTR", "IPSWICH", "STWMARKT", "DISS","NRCH"
]
df['tpl'] = pd.Categorical(df['tpl'], categories=custom_order, ordered=True)
df = df.sort_values(by=['rid', 'tpl'])
df.head(30)
```

```python

df_not_livs = df[df['tpl'] != 'NRCH']

df_livs = df[df['tpl'] == 'NRCH']
df = pd.concat([df_not_livs, df_livs])
df.head(30)
```

```python
# Empty those columns as we do not need them for training our model
empty_columns = [
    'wtp', 'arr_et', 'arr_wet', 'pass_et', 'pass_wet', 'arr_atRemoved', 
    'pass_atRemoved', 'dep_et', 'dep_wet', 'dep_atRemoved', 'pass_at'
]
df[empty_columns] = np.nan
```

```python
empty_rows = df.isna().all(axis=1)
```

```python
clean_df = df.dropna(how='all', subset=['pta', 'ptd', 'wta', 'wtd', 'arr_at', 'dep_at'])
```
